```python
x=df.iloc[:,2:].values
x
```

```python
scale=MinMaxScaler()
features=scale.fit_transform(x)
features
```

Elbow method to get optimal value

```python
inertia=[]
for i in range(1,16):
  k_means=KMeans(n_clusters=i)
  k_means=k_means.fit(features)
  inertia.append(k_means.inertia_)
```

```python
#Elbow graph
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16),inertia,'bo-')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
```

```python
#elbow visualizer
plt.subplot(1,2,2)
kmeans=KMeans()
visualize=KElbowVisualizer(kmeans,k=(1,16))
visualize.fit(features)
plt.suptitle("elbow graph and visualizer")
visualize.poof()
plt.show
```

silhoutte score for each k value

```python
silhoutte_avg= []
for i in range(2,16):
  #initialize kmeans cluster
  k_means=KMeans(n_clusters=i)
  cluster_labels=k_means.fit_predict(features)
#silhoutte score
  silhoutte_avg.append(silhouette_score(features,cluster_labels))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(2,16),silhoutte_avg,'bX-')
plt.xlabel("Values of K")
plt.ylabel("Silhoutte score")
plt.title("Silhoutte analysis for optimal k")
plt.show()
```

KMeans Model
Here we will take K value as 3 as per Elbow Method

```python
model=KMeans(n_clusters=3)
model=model.fit(features)
```

```python
y_km=model.predict(features)
centers=model.cluster_centers_
```

```python
df['Cluster']=pd.DataFrame(y_km)
df.to_csv("Cluster_data",index=False)
```
